### Multiple Linear Regression

<ul>
    <li>Same as simple linear regression, except with multiple predictors.</li>
    <li>There are however, different metrics when using multiple linear regression as opposed to simple linear regression.</li>
</ul>


<b>Degrees of Freedom</b>
<ul>
    <li>The no. of values that are free to vary in the calculation of a statistic.</li>
    <li>df = n-p-1, where p = no. of variables, n = no. of data points</li>
    <li>For example, in linear regression, when we have 2 points, we say that we have 0 degrees of freedom. This is because 2-1-1=0. This is because our line must pass between these 2 points, and we do not freedom to draw any other line.</li>
    <li> When we have three points, a we draw a best fit line between 2 points, but we realize that our <b>line has the freedom to move between the 2 points and the point that the line does not pass through (blue line)</b>, but it will never be on the side with no points (red line) </li>
</ul>

<div>
    
</div>

<b>Adjusted R-Squared</b>
<ul>
    <li>R-squared = normal R-squared. As we increase the no. of variables, the degree of variables decrease. However, as we add more variables, R-Squared increases, causing it to become closer to 1. This means that when we add a variable that does not bring about significant increase in R-Squared, numerator remains constant, while denominator becomes smaller. Which causes overall adjusted R-Squared to become smaller. </li>
    <li>Adjusted R squared punishes the addition of not so important variables. </li>
</ul>

<div>
    
</div>

<b>Linear Relationship/ Multicollinearity</b>
<ul>
    <li>Linear Relationship: Each of the independent and dependent features have a linear relationship. Can be checked with scatter plots or Pearson Correlation Coefficient. Does not matter if it negative or positive. <b>For variables with non-linear relationship, we drop them, but if there is a non-linear relationship, we might need to use a log or exponential transform to make it linear.</b></li>
    <li>Multicollinearity: When 2 indepdent features have a linear relationship/ highly correlated, we call it multicollinearity. In this case, we only need to include one of these features in our model and including the other one is redundant. To check this, we use the correlation matrix.</li>
</ul>

<b>Autocorrelation</b>
<ul>
    <li>When the value of a variable of one record is dependent on the value of the same variable in another record. For example, time series data for stock data. The stock prices of tomorrow depends on the stock price of today.</li>
    <li>We can measure the autocorrelation of a feature between time periods by subbing in the values into the autocorrelation formula</li>
    <li><b>To handle rows with autocorrelation, we can use the sliding window approach, where we shift the values in the previous time periods as seperate features. This effectively transforms the data into a normal regression problem</b></li>
    <li>For example, for k=4, we shift the previous 4 periods into one row, see for t0=84, we change 56,36,14,8 into seperate columns.</li>
    <li> <b>We can visualize the correlation between time periods using the autocorrelation plot. </b></li>
</ul>

<b>Endogeneity</b>
<ul>
    <li>"Include the variables that need to be included"</li>
    <li>Our data is endogenic when a feature is correlated with the error term. (i.e. when we can express our error as some function of x)</li>
    <li>If endogeneity is present in the between the error term and a feature, it means that we should be including that feature in our model.</li>
    <li> To counter this, we can include all the features and use adjusted R-squared or feature selection to prune out the unwanted vars</li>
</ul>

<div>
    
</div>

<b>Normality of residuals</b>
<ul>
    <li>Residual = error between predicted value and actual value.</li>
    <li>Assumption of normality: When we plot the distribution of all the residuals, it should be normal. (The mean of the errors should be zero, majority of the observation is centered around the mean and 68% of the obs are within 2 stdevs away from the mean.)</li>
    <li>If this assumption is not true, it means that multiple linear regression model is not a good fit and we need to use a different regression (for example, polynomial regression)</li>
</ul>

<div>
    
</div>

<b>Dummy variable trap</b>
<ul>
    <li>When we hot-encode labelled data, we only include (n-1) where n = no. of categories.</li>
    <li>This is because if we include all 3, we will get <b>perfect </b>multicollinearity, which will cause the error calculation to fail.</li>
</ul>


<b> Multiple Linear Regression Assumptions </b>

<ul>
    <li>There cannot be multicollinearity or autocorrelation between variables</li>
    <li> There should not be endogeneity in the variable</li>
    <li>Sample size should be large enough (20 samples per var) i,e: there should be at least 20 observations for each variable. For example, if we have 5 variables, we should have min 100 records</li>
     <li>Error needs to have normality</li>
</ul>

```python
# Plotting autocorrelation plot

# Notice perfect 1 collinearity at t = t0
# The values of collinearity between different time periods can be seen based on x axis (time period)
# and y axis (correlation)

import pandas as pd
import matplotlib.pyplot as plt
f = pd.read_csv("./data/03+-+corr.csv")
f['t0'] = pd.to_numeric(f['t0'],downcast='float');

plt.style.use('ggplot')
plt.xlabel("Time period")
plt.ylabel("Correlation")
plt.acorr(f['t0'],maxlags=10);
```
